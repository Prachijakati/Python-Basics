# Problem 1: Mobile Recharge Validation System


```python
def validate_recharge(amount, plan):
    """
    Function to validate mobile recharge request.
    Rules:
    - Amount must be >= 50
    - Plan must be one of the valid plans
    """
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("âŒ Recharge amount must be at least â‚¹50.")
        return False

    if plan not in valid_plans:
        print("âŒ Invalid plan selected.")
        return False

    print("âœ… Recharge Successful!")
    return True


# Retry using while loop
while True:
    try:
        amount = int(input("Enter recharge amount: â‚¹"))
        plan = int(input("Enter recharge plan (199, 299, 399, 599): "))

        if validate_recharge(amount, plan):
            break
        else:
            print("Please try again.\n")

    except ValueError:
        print("âŒ Invalid input. Please enter numbers only.\n")
```

# Problem 2: Inventory Reorder Alert System


```python
def check_inventory(stock_dict):
    """
    Function to check stock levels.
    Prints Reorder Alert if stock < 15
    """

    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"ðŸ”´ {product}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"ðŸŸ¢ {product}: Stock OK (Stock = {quantity})")


# Sample dictionary
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 8,
    "Monitor": 20
}

check_inventory(inventory)
```

# Problem 3: Student Result Processing System



```python
def process_results(marks_list):
    """
    Function to calculate average and return Pass/Fail.
    """

    total = 0

    # Loop to calculate total
    for marks in marks_list:
        total += marks

    average = total / len(marks_list)

    print(f"Average Marks: {average:.2f}")

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Example
marks = [60, 45, 70, 55]
result = process_results(marks)
print("Result:", result)
```

# Problem 4: Cab Fare Estimator



```python
def calculate_fare(distance, peak):
    """
    Function to calculate cab fare.
    Base fare = â‚¹50
    â‚¹12 per km
    Peak hour = 25% extra
    """

    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak.lower() == "yes":
        fare += fare * 0.25

    return fare


# Retry using while loop
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(distance, peak)

        print(f"ðŸš– Total Fare: â‚¹{total_fare:.2f}")

        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            break

    except ValueError:
        print("âŒ Invalid input. Please try again.\n")
```

# Problem 5: Attendance Eligibility Checker



```python
def check_eligibility(attendance_list):
    """
    Function to calculate attendance percentage
    and return eligibility.
    """

    total_days = len(attendance_list)
    present_days = 0

    # Count present days
    for day in attendance_list:
        if day.upper() == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100
    print(f"Attendance Percentage: {percentage:.2f}%")

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example
attendance = ["P", "A", "P", "P", "A", "P", "P"]
status = check_eligibility(attendance)
print("Status:", status)
```

# Problem 6: Password Strength Checker



```python
def check_password(password):
    """
    Function to check password strength.
    Rules:
    - Minimum length = 8
    - At least one digit
    - At least one special character (@#$)
    """

    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    has_digit = False
    has_special = False

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include digit and special character @#$)"


# Example
user_password = input("Enter password: ")
print(check_password(user_password))
```
