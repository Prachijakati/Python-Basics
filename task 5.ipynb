# Problem 1: Post Engagement Analyzer



```python
def analyze_engagement(likes_list):
    """
    Calculates total likes and determines post status.
    """

    total_likes = 0

    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")


# Example
likes = [200, 150, 300, 400]
analyze_engagement(likes)
```

# Problem 3: Rainfall Adequacy Checker



```python
def rainfall_analysis(rainfall_list, required_level):
    """
    Calculates average rainfall and checks adequacy.
    """

    total = 0

    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print(f"Average Rainfall: {average}")
    print(f"Rainfall Status: {status}")


# Example
rainfall = [70, 75, 68, 80]
rainfall_analysis(rainfall, 70)
```

# Problem 4: Duplicate Account Detection



```python
def detect_duplicates(usernames):
    """
    Detects duplicate usernames using set.
    """

    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example
users = ["john", "alice", "john", "mark"]
detect_duplicates(users)
```

# Problem 6: Premium Crop Price Filter



```python
def filter_premium_crops(prices):
    """
    Filters crop prices above ₹2000.
    """

    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print(f"Premium Crops: {premium}")


# Example
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

# Problem 7: Application Health Checker

# Problem 9: Student Attendance Eligibility System



```python
def attendance_eligibility(attendance_list):
    """
    Calculates attendance percentage and checks eligibility.
    """

    total_days = len(attendance_list)
    present_days = 0

    for day in attendance_list:
        if day.upper() == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print(f"Attendance Percentage: {percentage}")
    print(f"Exam Eligibility: {status}")


# Example
attendance = ["P", "P", "A", "P", "P"]
attendance_eligibility(attendance)
```

# Problem 10: Smart Electricity Bill Analyzer



```python
def calculate_electricity_bill(units):
    """
    Calculates electricity bill based on slab rates
    and classifies usage.
    """

    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


# Example
bill_amount, usage_status = calculate_electricity_bill(250)

print(f"Total Bill: ₹{bill_amount}")
print(f"Usage Level: {usage_status}")
```
