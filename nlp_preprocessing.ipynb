"Love" and "love" are treated as different tokens.
This increases vocabulary size unnecessarily.
That’s why we convert text to lowercase.

"Love" → different token

"love" → different token

4. Stemming vs Lemmatization
Stemming and lemmatization are both techniques used in text processing, but they differ in how they work.

Stemming is a rule-based method that removes word endings or suffixes using simple heuristics. The output is not always a real or valid word. For example, "running" is reduced to "run", but in some cases the result may be less meaningful. Also, a word like "better" remains "better" because stemming does not consider context or meaning.

Lemmatization, on the other hand, is a dictionary-based method that uses vocabulary and morphological analysis. It always produces a valid, meaningful word. For example, "running" becomes "run", and "better" is converted to "good" because it understands the word’s base form and context.

Task 2 - Full preprocessing function

```python
import re

def preprocess_text(text):
    # Error handling
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove emojis & special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Handle repeated characters (soooo → soo)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short words except 'no', 'not'
    cleaned_tokens = []
    for word in tokens:
        if len(word) > 2 or word in ['no', 'not']:
            cleaned_tokens.append(word)

    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence
```

Task 3 - Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://google.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    
    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
    print("-"*50)
    
    results.append(tokens)
```

Task 4 - Token analytics

```python
import numpy as np

for i, tokens in enumerate(results):
    if len(tokens) == 0:
        continue
        
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(word) for word in tokens])

    print(f"Sentence {i+1}")
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Avg Token Length:", round(avg_length, 2))
    print("-"*40)
```

Most noisy sentence:
"Win $$$ now!!! Limited offer!!!"

Contains symbols, noise, repetition

Most meaningful after cleaning:
"I am not happy with this"

Keeps sentiment structure intact

Task 5 - Frequency analysis

```python
from collections import Counter

all_tokens = []
for tokens in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 frequent words:")
print(counter.most_common(10))

print("\nLeast 5 frequent words:")
print(counter.most_common()[-5:])
```

Task 6 - Full pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
